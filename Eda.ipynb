```python
df.dropna(inplace=True)
```

```python

species_counts = df.groupby('island')['species'].value_counts().unstack().fillna(0)
species_counts = species_counts.reset_index().melt(id_vars='island', value_name='count', var_name='species')
species_counts
```

```python
df['species'].value_counts().plot(kind='bar',colormap='viridis')
```

```python
color_palette = ['blue', 'black', 'red']
sns.catplot(x='species',hue='island',kind='count',data=df,palette=color_palette)


```

which island has which  sex type of penguin

Which specie has more body weight hue sex

```python
sns.barplot(x=df['species'],y=df['body_mass_g'],data=df,hue='sex')
plt.title('Body Mass by Species and Sex')
plt.show()
```

```python
sns.relplot(x='flipper_length_mm',y='body_mass_g',kind='scatter',data=df,hue='sex')
plt.title('Flipper length vs weight')
plt.figure(figsize=(10,5))
plt.show()
```
